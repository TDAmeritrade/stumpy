# Intro

In this notebook, we would like to derive the eq(2) of the paper [VALMOD](https://arxiv.org/pdf/2008.13447.pdf).

**Notation:** $T_{i,m} = T[i:i+m]$, a subsequence of `T` that starts at index `i` and has length `m`. 

**The idea goes as follows:** <br>
"Given the distance profile of $T_{j,m}$, how can we find a lower bound for distance profile of $T_{j,m+k}$", where $T_{j,m+k}$ represents a sequence that starts from the same index `j` with length `m+k`? In other words, can we find **Lower Bound (LB)** for $d(T_{j,m+k}, T_{i,m+k})$ only by help of $T_{j,m}$, $T_{i,m}$, and $T_{j,m+k}$? (So, the last `k` elements of $T_{i,m+k}$ are unknown)

## Non-normalized distance


$$
\begin{align}
    d^{(m+k)}_{j,i} ={}& 
        \sqrt[\leftroot{5}\uproot{5}p]{
                \sum\limits_{t=0}^{m+k-1}{
                \bigg\lvert{
                 T[i+t] - T[j+t]
                 }\bigg\rvert
                }^{p}
                }
    \\
    ={}&
    \sqrt[\leftroot{5}\uproot{5}p]{
        \sum\limits_{t=0}^{m-1}{
            \bigg\lvert{
            T[i+t] - T[j+t]
            }\bigg\rvert
         }^{p}
         +
         \sum\limits_{t=m}^{m+k-1}{
            \bigg\lvert{
            T[i+t] - T[j+t]
            }\bigg\rvert
         }^{p}
      }
    \\
    \geq{}&
    \sqrt[\leftroot{5}\uproot{5}p]{
        \sum\limits_{t=0}^{m-1}{
            \bigg\lvert{
            T[i+t] - T[j+t]
            }\bigg\rvert
         }^{p}
      }
    \\
\end{align}
$$


Therefore:


$$
\begin{align}
    d^{(m+k)}_{j,i} \geq{}&
    d^{(m)}_{j,i}
\end{align}
$$


In other words, we can simply use the p-norm distance between $T_{i,m}$ and $T_{j,m}$ as the lower-bound value for the distance between $T_{i,m+k}$ and $T_{j,m+k}$.

## Normalized distance

In z-normalized distance, one should note that $d^{(m+k)}_{j,i} \geq d^{(m)}_{j,i}$ is not necessarily correct. In other words, the distance between two subsequences does not necessarily increase by making them longer. However, it seems there is a very nice relationship between $d_{j,i}^{(m)}$ and the lower-bound value of $d_{j,i}^{(m+k)}$.

### Derving Equation (2)


$$
\begin{align}
    d^{(m+k)}_{j,i} ={}& 
        \sqrt{\sum\limits_{t=0}^{m+k-1}{{
        \left(\frac{T[i+t] - \mu_{i,m+k}}{\sigma_{i,m+k}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m+k}}\right)
        }^{2}}} 
    \\
    d^{(m+k)}_{j,i} ={}& 
        \sqrt{
        \sum\limits_{t=0}^{m-1}{{
        \left(\frac{T[i+t] - \mu_{i,m+k}}{\sigma_{i,m+k}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m+k}}\right)
        }^{2}}
        +
        \sum\limits_{t=m}^{m+k-1}{{
        \left(\frac{T[i+t] - \mu_{i,m+k}}{\sigma_{i,m+k}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m+k}}\right)
        }^{2}}
        } 
    \\
    \geq{}&
        \sqrt{\sum\limits_{t=0}^{m-1}{{
        \left(\frac{T[i+t] - \mu_{i,m+k}}{\sigma_{i,m+k}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m+k}}\right)
        }^{2}}}
    \\
\end{align}
$$


So, the Lower-Bound (LB) value for $d_{j,i}^{(m+k)}$ can be obtained by minimizing the right-hand side:


$$
\begin{align}
    LB ={}& 
        \min \sqrt{\sum\limits_{t=0}^{m-1}{{
        \left(\frac{T[i+t] - \mu_{i,m+k}}{\sigma_{i,m+k}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m+k}}\right)
        }^{2}}} 
    \\
    ={}&
    \min \sqrt{\sum\limits_{t=0}^{m-1}{{
        \left[\frac{1}{\sigma_{j,m+k}}
            \left(
            \frac{T[i+t] - \mu_{i,m+k}}{\frac{\sigma_{i,m+k}}{\sigma_{j,m+k}}} - \frac{T[j+t] - \mu_{j,m+k}}{1}
            \right)
        \right]
        }^{2}}}
    \\
    ={}&
    \min \sqrt{
        \sum\limits_{t=0}^{m-1}{{
            \left[
                \frac{\sigma_{j,m}}{\sigma_{j,m}}
                \frac{1}{\sigma_{j,m+k}}
                \left(
                     \frac{T[i+t] - \mu_{i,m+k}}{\frac{\sigma_{i,m+k}}{\sigma_{j,m+k}}} 
                     - 
                     \frac{T[j+t] - \mu_{j,m+k}}{1}
                 \right)
            \right]
                }^{2}
        }
        }
    \\
    ={}&
    \min \sqrt{
        \sum\limits_{t=0}^{m-1}{{
            \left[
                \frac{\sigma_{j,m}}{\sigma_{j,m+k}}
                \left(
                     \frac{T[i+t] - \mu_{i,m+k}}{\sigma_{j,m}\frac{\sigma_{i,m+k}}{\sigma_{j,m+k}}} 
                     - 
                     \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m}}
                 \right)
            \right]
                }^{2}
        }
        }
    \\
    ={}&
    \frac{\sigma_{j,m}}{\sigma_{j,m+k}} \times \min \sqrt{\sum\limits_{t=0}^{m-1}{\left(\frac{T[i+t] - \mu_{i,m+k}}{\frac{\sigma_{j,m} \sigma_{i,m+k}}{\sigma_{j,m+k}}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m}}\right)^{2}}} \quad(1)
    \\
\end{align}
$$


**Note:** that the unknown variables are $\mu_{i,m+k}$ and $\sigma_{i,m+k}$. Also, note that all $\mu$ and $\sigma$ values are **constant** regardless of them being known or unknown. <br>

We subtitute $\mu_{i,m+k}$ with $\mu^{'}$, and $\frac{\sigma_{j,m} \sigma_{i,m+k}}{\sigma_{j,m+k}}$ with $\sigma^{'}$. Note that the unknown variables are now $\mu^{'}$ and $\sigma^{'}$. <br>

Also, we define $\alpha_{t}$ as:


$$
\begin{align}
    \alpha_{t} \triangleq{}& 
        {
        \frac{T[i+t] - \mu^{'}}{\sigma^{'}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m}} \quad (2)
        } 
    \\
\end{align}
$$


Therefore, we have:


$$
\begin{align}
    LB ={}& 
        \frac{\sigma_{j,m}}{\sigma_{j,m+k}}
        \sqrt{\min \sum \limits_{t=0}^{m-1} {\alpha_t^{2}}} 
    \\
    ={}&
    \frac{\sigma_{j,m}}{\sigma_{j,m+k}}
        \sqrt{\min f(\mu^{'}, \sigma^{'})} \quad (3)
    \\
    f(\mu^{'}, \sigma^{'}) ={}&
    \sum \limits_{t=0}^{m-1} {\alpha_t^{2}} \quad (4)
    \\
\end{align}
$$


**To find extrema points, we first need to find the critical point(s) by solving the single system of equations below.**  In other words, we are looking for $\mu^{'}$ and $\sigma^{'}$ that satisfies both equations below:




$$
\begin{align}
    \frac{\partial{f}}{\partial{\mu^{'}}} = 0 \quad (5)
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} = 0 \quad (6)
    \\
\end{align}
$$


**Solving $\frac{\partial{f}}{\partial{\mu^{'}}} = 0$:**


$$
\begin{align}
    \frac{\partial{f}}{\partial{\mu^{'}}} ={}& 
    \sum \limits_{t=0}^{m-1}{
        \frac{\partial{(\alpha_{t}^{2})}}{\partial{\mu^{'}}}
    }
    \\
    \frac{\partial{f}}{\partial{\mu^{'}}} ={}& 
    \sum \limits_{t=0}^{m-1}{
        2\frac{\partial{(\alpha_{t})}}{\partial{\mu^{'}}}\alpha_{t}
    }
    \\
    \frac{\partial{f}}{\partial{\mu^{'}}} ={}&
    \sum \limits_{t=0}^{m-1} {
    2\left(
    \frac{-1}{\sigma^{'}}
    \right)
    \alpha_{t}} 
    \\
    0 ={}&
    \frac{-2}{\sigma^{'}}\sum \limits_{t=0}^{m-1}{\alpha_{t}}
    \\
\end{align}
$$


Please note that $\sigma^{'}$ is constant and thus it was factered out of the summation. <br>
This gives us:


$$
\begin{align}
    \sum \limits_{t=0}^{m-1}{\alpha_{t}} = 0 \quad (7)
\end{align}
$$


**Exapanding (7):**


$$
\begin{align}
    \sum \limits_{t=0}^{m-1} \alpha_{t} ={}& 
    0
    \\
    \sum \limits_{t=0}^{m-1} {\frac{T[i+t] - \mu^{'}}{\sigma^{'}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m}}} ={}& 
    0
    \\
    \frac{1}{\sigma^{'}}\left(\sum \limits_{t=0}^{m-1}T[i+t] - \sum \limits_{t=0}^{m-1} \mu^{'}\right) - 
    \frac{1}{\sigma_{j,m}}\left(\sum \limits_{t=0}^{m-1}T[j+t] - \sum \limits_{t=0}^{m-1} \mu_{j,m+k}\right) ={}& 
    0
    \\
    \frac{1}{\sigma^{'}}\left(m\mu_{i,m} - m\mu^{'}\right) - 
    \frac{1}{\sigma_{j,m}}\left(m\mu_{j,m} - m\mu_{j,m+k}\right) ={}& 
    0
    \\
    \sigma_{j,m}\left(\mu_{i,m} - \mu^{'}\right) - 
    \sigma^{'}\left(\mu_{j,m} - \mu_{j,m+k}\right) ={}& 
    0
    \\
    \sigma_{j,m} \mu^{'} + 
    \left(\mu_{j,m} - \mu_{j,m+k}\right)\sigma^{'} - \sigma_{j,m}\mu_{i,m} ={}& 
    0 \quad (8)
\end{align} 
$$


**Solving $\frac{\partial{f}}{\partial{\sigma^{'}}} = 0$:**


$$
\begin{align}
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}& 
    \sum \limits_{t=0}^{m-1}{
        \frac{\partial{(\alpha_{t}^{2})}}{\partial{\sigma^{'}}}
    }
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}& 
    \sum \limits_{t=0}^{m-1}{
        2\frac{\partial{(\alpha_{t})}}{\partial{\sigma^{'}}}\alpha_{t}
    }
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}&
    \sum \limits_{t=0}^{m-1} {
    2 \left(
        \frac{-\left({T[i+t] - \mu^{'}}\right)}{\sigma^{'2}}
    \right)
    \alpha_{t}} 
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}&
    \frac{-2}{\sigma^{'2}}\sum \limits_{t=0}^{m-1}{\left({T[i+t] - \mu^{'}}\right) \alpha_{t}}
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}&
    \frac{-2}{\sigma^{'2}}\sum \limits_{t=0}^{m-1}{\left({T[i+t]\alpha_{t} - \mu^{'}\alpha_{t}}\right)}
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}&
    \frac{-2}{\sigma^{'2}}
    {\left(
    \sum \limits_{t=0}^{m-1}{T[i+t]\alpha_{t}} 
    - 
    \sum \limits_{t=0}^{m-1}{\mu^{'}\alpha_{t}}
    \right)
    }
    \\
    \frac{\partial{f}}{\partial{\sigma^{'}}} ={}&
    \frac{-2}{\sigma^{'2}}
    {\left(
    \sum \limits_{t=0}^{m-1}{T[i+t]\alpha_{t}} 
    - 
    \mu^{'}\sum \limits_{t=0}^{m-1}{\alpha_{t}}
    \right)
    }
    \\
    0 ={}&
    \frac{-2}{\sigma^{'2}}
    {\left(
    \sum \limits_{t=0}^{m-1}{T[i+t]\alpha_{t}} 
    - 
    \mu^{'}\cdot 0
    \right)
    }
    \\
    0 ={}&
    \frac{-2}{\sigma^{'2}}
    {
    \sum \limits_{t=0}^{m}{T[i+t]\alpha_{t}} 
    }
\end{align}
$$


Note: In the calculations above, we substituted 0 for  $\sum \limits_{t=0}^{m-1}{\alpha_{t}}$ according to eq(7).

And, this gives:


$$
\begin{align}
    \sum \limits_{t=0}^{m-1}{T[i+t]\alpha_{t}}  ={}&
    0 \quad (9)
\end{align}
$$


**Expanding (9):**


$$
\begin{align}
    \sum \limits_{t=0}^{m-1} T[i+t] \left(\frac{T[i+t] - \mu^{'}}{\sigma^{'}} - \frac{T[j+t] - \mu_{j,m+k}}{\sigma_{j,m}}\right) = 0
    \\
\end{align}
$$



$$
\begin{align}

Now, recall that the pearson correlation $\rho_{ij}$ between two subsequences starting at locations $i$ and $j$, respectively, and both of length $m$ is defined as follows:

$$
\begin{align}
\rho_{ij} ={}&
    \frac{
    COV(T_{i,m}T_{j,m})}{
    \sigma_{i,m}\sigma_{j,m}
    }
    \\
    ={}&
    \frac{
    E\left[
    (T_{i,m} - \mu_{i,m})(T_{j,m} - \mu_{j,m})
    \right]}
    {
    \sigma_{i,m}\sigma_{j,m}
    }
    \\
    ={}&
    \frac{
    \frac{1}{m}\sum\limits_{t=0}^{m-1}
    (T[i+t] - \mu_{i,m})(T[j+t] - \mu_{j,m})
    }
    {
    \sigma_{i,m}\sigma_{j,m}
    }
    \\
    ={}&
    \frac{
    \sum\limits_{t=0}^{m-1}
    T[i+t]T[j+t] 
    -
    \sum\limits_{t=0}^{m-1}
    \mu_{i,m}T[j+t]
    -
    \sum\limits_{t=0}^{m-1}
    \mu_{j,m}T[i+t]
    +
    \sum\limits_{t=0}^{m-1}\mu_{i,m}\mu_{j,m}
    }{
    m\sigma_{i,m}\sigma_{j,,m}
    }
    \\
    ={}&
    \frac{
    \sum\limits_{t=0}^{m-1}
    T[i+t]T[j+t] 
    -
    \mu_{i,m}\sum\limits_{t=0}^{m-1}
    T[j+t]
    -
    \mu_{j,m}\sum\limits_{t=0}^{m-1}
    T[i+t]
    +
    \sum\limits_{t=0}^{m-1}\mu_{i,m}\mu_{j,m}
    }{
    m\sigma_{i,m}\sigma_{j,m}
    }
    \\
    ={}&
    \frac{
    \sum\limits_{t=0}^{m-1}
    T[i+t]T[j+t] 
    -
    \mu_{i,m}\cdot m\mu_{j,m}
    -
    \mu_{j,m}\cdot m\mu_{i,m}
    +
    m\mu_{i,m}\mu_{j,m}
    }{
    m\sigma_{i,m}\sigma_{j,m}
    }

    ={}&
    \frac{
    \sum\limits_{t=0}^{m-1}

    -
    m\mu_{i,m}\mu_{j,m}
    }{
    m\sigma_{i,m}\sigma_{j,m}
    }
    \\
\end{align}
$$

    \mu_{i,m}\mu_{j,m}\sigma^{'} + \mu_{i,m}\mu_{j,m+k}\sigma^{'} 
    + \sigma_{j,m}\mu_{i,m}^{2} +
    \mu_{i,m}\sigma_{j,m}\mu^{'} + \rho_{ij}\sigma_{i,m}\sigma_{j,m}\sigma^{'} + \mu_{i,m}\mu_{j,m}\sigma^{'} - \mu_{i,m}\mu_{j,m+k}\sigma^{'} - \sigma_{j,m}\mu_{i,m}^{2} - \sigma_{j,m}\sigma_{i,m}^{2}
    ={}& 0
    \\
    \rho_{ij}\sigma_{i,m}\sigma_{j,m}\sigma^{'} - \sigma_{j,m}\sigma_{i,m}^{2} 
    ={}& 0
    \\
    \sigma_{i,m}\sigma_{j,m}
    \left(
    \rho_{ij}\sigma^{'} - \sigma_{i,m}
    \right)
    ={}&
    0
    \\
\end{align}
$$


Hence:


$$
\begin{align}
    \sigma^{'} = \frac{\sigma_{i,m}}{\rho_{ij}} \quad (11)
\end{align}
$$


Note that we assumed $\sigma_{i,m}$ and $\sigma_{j,m}$ cannot be zero. Also, since standard deviations are positive, eq(11) is valid only if $\rho_{ij} \gt 0$.

And, subsituting eq(11) in eq(8):


$$
\begin{align}
\sigma_{j,m} \mu^{'} + 
    \left(\mu_{j,m} - \mu_{j,m+k}\right)(\frac{\sigma_{i,m}}{\rho_{ij}}) - \sigma_{j,m}\mu_{i,m} 
    ={}& 0 
    \\
    \frac{1}{\sigma_{j,m}}\left[
    \sigma_{j,m} \mu^{'} + 
    \left(\mu_{j,m} - \mu_{j,m+k}\right)(\frac{\sigma_{i,m}}{\rho_{ij}}) - \sigma_{j,m}\mu_{i,m} 
    \right]
    ={}& 0 
    \\
    \mu^{'} + \left(\mu_{j,m} - \mu_{j,m+k}\right)(\frac{\sigma_{i,m}}{\rho_{ij}\sigma_{j,m}}) - \mu_{i,m} 
    ={}& 0 
\end{align}
$$


Hence:


$$
\begin{align}
    \mu^{'} =  \mu_{i,m} - \frac{\sigma_{i,m}}{\rho_{ij}\sigma_{j,m}} \left(\mu_{j,m} - \mu_{j,m+k}\right) \quad(12)
\end{align}
$$
           

**Therefore, the critical point of function $f(\mu^{'},\sigma^{'})$ is:**


$$
\begin{align}

    \frac{\sigma_{i,m}}{\rho_{ij}} \quad (11)

\end{align}

        \frac{
            \sigma_{i,m}^{2} + (\mu_{i,m}-\mu^{'})^{2}

            \sigma^{'2}

            + \mu^{'}\mu_{j,m+k}

            }

        ={}&
        \frac{
            \sigma_{i,m}^{2} + (\mu_{i,m}-\mu^{'})^{2}

            \sigma^{'2}
            }

         -

            - \mu_{j,m+k})

            \sigma_{i,m}^{2}

           \frac{\mu_{j,m}
            - \mu_{j,m+k}}{\sigma_{j,m}})

           \frac{\mu_{j,m}
            - \mu_{j,m+k}}{\sigma_{j,m}})
            \right]

           ={}&
        \frac{
            \sigma_{i,m}^{2}

           \\
           \geq{}&
           2\frac{(-\rho_{ij})\sigma_{i,m}}{\sigma^{'}} \quad (19)

           \geq{}&

\end{align}
$$
